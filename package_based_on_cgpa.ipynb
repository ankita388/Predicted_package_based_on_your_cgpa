```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import joblib

# Step 1: Load and clean the data
df = pd.read_csv("package.csv")
df.columns = df.columns.str.strip()  # Remove spaces just in case

# Step 2: Rename 'PACKAGE(LPM)' to 'Package' for simplicity
df.rename(columns={'PACKAGE(LPM)': 'Package'}, inplace=True)

# Step 3: Prepare the features and labels
X = df[['CGPA']]
y = df['Package']

# Step 4: Train the model
model = LinearRegression()
model.fit(X, y)

# Step 5: Save the model
joblib.dump(model, 'model.pkl')

print("✅ Model trained and saved as model.pkl")
```
